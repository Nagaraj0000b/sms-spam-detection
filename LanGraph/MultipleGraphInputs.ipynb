```python
from typing import Dict,TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    values: list[int]
    name: str
    result:str
```

```python
def process_values(state:AgentState)->AgentState:
    print(state)
    state["result"]="Hey "+state["name"]+" the sum is "+str(sum(state["values"]))
    print(state)
    return state
```

```python
graph=StateGraph(AgentState)
graph.add_node("ok", process_values)
graph.set_entry_point("ok")
graph.set_finish_point("ok")
app=graph.compile()
```

```python
from IPython.display import Image,display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
answer=app.invoke({"values":[1,2,3,4,5],"name":"harold"})


```
